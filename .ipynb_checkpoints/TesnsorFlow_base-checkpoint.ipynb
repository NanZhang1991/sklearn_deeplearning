è¯»è€…åœ¨çœ‹åˆ°æ¯ç§å¼ é‡æ—¶ï¼Œèƒ½å¤Ÿç›´è§‚åœ°è”æƒ³åˆ°å®ƒä¸»è¦çš„ç‰©ç†æ„ä¹‰å’Œç”¨é€”ï¼Œå¯¹åç»­å¼ é‡çš„ç»´åº¦

### æ ‡é‡
åœ¨ TensorFlow ä¸­ï¼Œæ ‡é‡æœ€å®¹æ˜“ç†è§£ï¼Œå®ƒå°±æ˜¯ä¸€ä¸ªç®€å•çš„æ•°å­—ï¼Œç»´åº¦æ•°ä¸º0ï¼Œshape ä¸º
[]ã€‚æ ‡é‡çš„å…¸å‹ç”¨é€”ä¹‹ä¸€æ˜¯è¯¯å·®å€¼çš„è¡¨ç¤ºã€å„ç§æµ‹é‡æŒ‡æ ‡çš„è¡¨ç¤ºï¼Œæ¯”å¦‚å‡†ç¡®åº¦(Accuracy,
acc)ï¼Œç²¾åº¦(Precision)å’Œå¬å›ç‡(Recall)ç­‰ã€‚
è€ƒè™‘æŸä¸ªæ¨¡å‹çš„è®­ç»ƒæ›²çº¿ï¼Œå¦‚å›¾ 4.1 æ‰€ç¤ºï¼Œæ¨ªåæ ‡ä¸ºè®­ç»ƒBatch æ­¥æ•°Stepï¼Œçºµåæ ‡åˆ†
åˆ«ä¸ºè¯¯å·®å˜åŒ–è¶‹åŠ¿(å›¾ 4.1(a))å’Œå‡†ç¡®åº¦å˜åŒ–è¶‹åŠ¿æ›²çº¿(å›¾ 4.1(b))ï¼Œå…¶ä¸­æŸå¤±å€¼loss å’Œå‡†ç¡®åº¦
å‡ç”±å¼ é‡è®¡ç®—äº§ç”Ÿï¼Œç±»å‹ä¸ºæ ‡é‡ã€‚

ä»¥å‡æ–¹å·®è¯¯å·®å‡½æ•°ä¸ºä¾‹ï¼Œç»è¿‡tf.keras.losses.mse(æˆ–tf.keras.losses.MSE)è¿”å›æ¯ä¸ªæ ·æœ¬
ä¸Šçš„è¯¯å·®å€¼ï¼Œæœ€åå–è¯¯å·®çš„å‡å€¼ä½œä¸ºå½“å‰batch çš„è¯¯å·®ï¼Œå®ƒæ˜¯ä¸€ä¸ªæ ‡é‡ï¼š

```python
out = tf.random.uniform([4,10]) #éšæœºæ¨¡æ‹Ÿç½‘ç»œè¾“å‡º
y = tf.constant([2,3,2,0]) # éšæœºæ„é€ æ ·æœ¬çœŸå®æ ‡ç­¾
y = tf.one_hot(y, depth=10) # one-hot ç¼–ç 
loss = tf.keras.losses.mse(y, out) # è®¡ç®—æ¯ä¸ªæ ·æœ¬çš„MSE
loss = tf.reduce_mean(loss) # å¹³å‡MSE
print(loss)
```

çš„æ•°å­¦è¿ç®—ç­‰æ“ä½œã€‚æœ¬èŠ‚æˆ‘ä»¬å°†åˆ©ç”¨æˆ‘ä»¬å·²ç»å­¦åˆ°çš„çŸ¥è¯†å»å®Œæˆä¸‰å±‚ç¥ç»ç½‘ç»œçš„å®ç°ï¼š
$$oğ‘¢ğ‘¡ = ğ‘Ÿğ‘’ğ‘™ğ‘¢{ğ‘Ÿğ‘’ğ‘™ğ‘¢{ğ‘Ÿğ‘’ğ‘™ğ‘¢[ğ‘‹@ğ‘Š1 + ğ‘1]@ğ‘Š2 + ğ‘2}@ğ‘Š + ğ‘ }$$
æˆ‘ä»¬é‡‡ç”¨çš„æ•°æ®é›†æ˜¯MNIST æ‰‹å†™æ•°å­—å›¾ç‰‡é›†ï¼Œè¾“å…¥èŠ‚ç‚¹æ•°ä¸º784ï¼Œç¬¬ä¸€å±‚çš„è¾“å‡ºèŠ‚ç‚¹æ•°æ˜¯
256ï¼Œç¬¬äºŒå±‚çš„è¾“å‡ºèŠ‚ç‚¹æ•°æ˜¯128ï¼Œç¬¬ä¸‰å±‚çš„è¾“å‡ºèŠ‚ç‚¹æ˜¯10ï¼Œä¹Ÿå°±æ˜¯å½“å‰æ ·æœ¬å±äº10 ç±»åˆ«
çš„æ¦‚ç‡ã€‚<br>
ã€€ã€€é¦–å…ˆåˆ›å»ºæ¯ä¸ªéçº¿æ€§å‡½æ•°çš„w,b å‚æ•°å¼ é‡ï¼š

```python
w1 = tf.Variable(tf.random.truncated_normal([784, 256], stddev=0.1))
b1 = tf.Variable(tf.zeros([256]))
w2 = tf.Variable(tf.random.truncated_normal([256, 128], stddev=0.1))
b2 = tf.Variable(tf.zeros([128]))
w3 = tf.Variable(tf.random.truncated_normal([128, 10], stddev=0.1))
b3 = tf.Variable(tf.zeros([10]))
```

åœ¨å‰å‘è®¡ç®—æ—¶ï¼Œé¦–å…ˆå°†shape ä¸º[ğ‘, 28,28]çš„è¾“å…¥æ•°æ®Reshape ä¸º[ğ‘, 784]:

```python
# [b, 28, 28] => [b, 28*28]
from tensorflow.keras import datasets
import os 
mnist_path = os.path.join(os.getcwd(),'data/mnist.npz')
(x,y),_= datasets.mnist.load_data(path=mnist_path)
x.shape, y.shape
x=tf.convert_to_tensor(x,dtype=tf.float32) / 255.
y=tf.convert_to_tensor(y,dtype=tf.int32)
print(x.shape)
x = tf.reshape(x, [-1, 28*28])
x.shape
```

å®Œæˆç¬¬ä¸€ä¸ªéçº¿æ€§å‡½æ•°çš„è®¡ç®—ï¼Œæˆ‘ä»¬è¿™é‡Œæ˜¾ç¤ºåœ°è¿›è¡ŒBroadcasting:

```python
# [b, 784]@[784, 256] + [256] => [b, 256] + [256] => [b, 256] + [b, 256]
h1 = x@w1 + tf.broadcast_to(b1, [x.shape[0], 256])
h1 = tf.nn.relu(h1)
h1.shape
```

åŒæ ·çš„æ–¹æ³•å®Œæˆç¬¬äºŒä¸ªå’Œç¬¬ä¸‰ä¸ªéçº¿æ€§å‡½æ•°çš„å‰å‘è®¡ç®—ï¼Œè¾“å‡ºå±‚å¯ä»¥ä¸ä½¿ç”¨ReLU æ¿€æ´»å‡½
æ•°ï¼š

```python
# [b, 256] => [b, 128]
h2 = h1@w2 + b2
print(h2.shape)
h2 = tf.nn.relu(h2)
# [b, 128] => [b, 10]
out = h2@w3 + b3
out.shape
```

å°†çœŸå®çš„æ ‡æ³¨å¼ é‡y è½¬å˜ä¸ºone-hot ç¼–ç ï¼Œå¹¶è®¡ç®—ä¸out çš„å‡æ–¹å·®ï¼š

```python
# mse = mean(sum(y-out)^2)
# [b, 10]
y_onehot = tf.one_hot(y, depth=10)
loss = tf.square(y_onehot - out)
# mean: scalar
loss = tf.reduce_mean(loss)
```

ä¸Šè¿°çš„å‰å‘è®¡ç®—è¿‡ç¨‹éƒ½éœ€è¦åŒ…è£¹åœ¨with tf.GradientTape() as tape ä¸Šä¸‹æ–‡ä¸­ï¼Œä½¿å¾—å‰å‘è®¡ç®—æ—¶
èƒ½å¤Ÿä¿å­˜è®¡ç®—å›¾ä¿¡æ¯ï¼Œæ–¹ä¾¿åå‘æ±‚å¯¼è¿ç®—ã€‚<br>
ã€€ã€€é€šè¿‡ tape.gradient()å‡½æ•°æ±‚å¾—ç½‘ç»œå‚æ•°åˆ°æ¢¯åº¦ä¿¡æ¯:

æŒ‰ç…§ï¼š
$$ ğœƒâ€² = ğœƒ - ğœ‚ âˆ— \frac{ğœ•â„’}{ğœ•ğœƒ} $$
æ¥æ›´æ–°ç½‘ç»œå‚æ•°

```python
train_db = tf.data.Dataset.from_tensor_slices((x,y)).batch(128)
train_iter = iter(train_db)   #è¿­ä»£å™¨
sample = next(train_iter)
print("batch: ", sample[0].shape, sample[1].shape)

#åˆ›å»ºæƒå€¼ï¼Œå®Œæˆå‰å‘ä¼ æ’­
#[batch,784] => [b, 256] => [b, 128] => [b, 10]
#w: [dim_in, dim_out]
#b: [dim_out]
w1 = tf.Variable(tf.random.truncated_normal([784, 256], stddev=0.1))#åŸæ¥å‡å€¼ä¸º0ï¼Œæ–¹å·®ç»´ä¸º1,ç°åœ¨æ–¹å·®å˜ä¸º0.1
b1 = tf.Variable(tf.zeros([256]))
w2 = tf.Variable(tf.random.truncated_normal([256, 128], stddev=0.1))
b2 = tf.Variable(tf.zeros([128]))
w3 = tf.Variable(tf.random.truncated_normal([128, 10], stddev=0.1))
b3 = tf.Variable(tf.zeros([10]))
# compute gradients
lr = 1e-3

#å‰å‘è¿ç®—


for epoch in range(10):         #iterate è¿­ä»£æ•´ä¸ªæ•°æ®é›†10æ¬¡ã€‚
    # å¤–å±‚for: å¯¹æ‰€æœ‰çš„å›¾ç‰‡åšå¾ªç¯ã€‚
    #for (x, y) in train_db:
    for step, (x, y) in enumerate(train_db):  #è¿­ä»£for every batch
        #x: [128, 28, 28]
        #y: [128]

        #ç»´åº¦å˜æ¢; [b, 28, 28]-> [b, 28*28]
        x = tf.reshape(x, [-1, 28*28])
        #x: [b, 28*28]
        #h1 = x@w1 + b1
        #[b, 784]@[784, 256] + [256] = [b, 256] + [256] => [b, 256] + [b, 256]

        #ä½¿ç”¨tensorflowè‡ªåŠ¨æ±‚å¯¼çš„è¿‡ç¨‹,å…¶ä¸­: w,bã€‚tf.GradientTape()å‚ä¸æ¢¯åº¦è®¡ç®—çš„ä»£ç æ”¾åˆ°è¿™é‡Œé¢ã€‚
        with tf.GradientTape() as tape:
            #GradientTapeé‡Œé¢é»˜è®¤åªä¼šè·Ÿè¸ªtf.Variable()ç±»å‹ã€‚å¦‚æœç±»å‹ä¸æ˜¯è¿™ä¸ªçš„è¯ã€‚è¿™é‡Œä¸ºtf.tensor,tf.Variableæ˜¯tf.tensorçš„ä¸€ç§ç‰¹æ®Šç±»å‹ã€‚
            #å› æ­¤ç®€å•çš„åŒ…è£…ä¸€ä¸‹ã€‚

            h1 = x@w1 + tf.broadcast_to(b1, [x.shape[0], 256]) #ç›´æ¥è‡ªåŠ¨å¹¿æ’­æœºåˆ¶ï¼Œä¹Ÿå¯ä»¥æ‰‹åŠ¨
            h1=tf.nn.relu(h1)
            #[b, 256] -> [b, 128]
            h2 = h1@w2 + b2
            h2=tf.nn.relu(h2)
            #[b, 128] -> [b, 10]
            out = h2@w3 + b3  #å¾—åˆ°å‰å‘è¾“å‡ºç»“æœã€‚

            #computer loss: è®¡ç®—è¯¯å·®å‡æ–¹å·®ã€‚
            # outç»´åº¦: [b, 10]
            # çœŸå®çš„y:  [b],ç»´åº¦è¿™é‡Œéœ€è¦æŠŠyå˜æˆä¸€ä¸ªone-hot ç¼–ç 
            y_onehot = tf.one_hot(y, depth=10)

            #mse = mean((y_onehot-out)^2)
            #shape : [b, 10]
            loss = tf.square(y_onehot - out)

            #mean: scalar
            #loss = tf.reduce_mean(loss) / b / 10 ä¸€èˆ¬æ¥è¯´é™¤ä»¥ä¸€ä¸ªbå°±å¤Ÿäº†ï¼Œæ¯ä¸ªbatchä¸Šçš„ä¸€ä¸ªå‡å€¼ã€‚å–å†³äºæ€ä¹ˆç†è§£ï¼Œéƒ½æ˜¯å¯ä»¥çš„ã€‚
            loss = tf.reduce_mean(loss) #è¿™é‡Œç›¸å½“äºä¸€ä¸ªæ”¾ç¼©ï¼Œæ­£å‘æ”¾ç¼©ä¸ä¼šå½±å“æ¢¯åº¦çš„æ–¹å‘ã€‚æ²¡æœ‰å½±å“çš„ã€‚
            #loss = tf._reduce_mean(tf.reduce_sum(loss,axis=1))#,æˆ‘è‡ªå·±çš„ç†è§£ã€‚

        #å¾—åˆ°ä¸€ä¸ªæ¢¯åº¦ã€‚éœ€è¦æ±‚è§£æ¢¯åº¦çš„æœ‰å“ªäº›å‘¢ï¼Ÿ
             #print(grads)  #ç»“æœï¼š[None, None, None, None, None, None]
        # w1 = w1 - lr* w1_grad, è¿™é‡Œä¸ºäº†ç»†èŠ‚ï¼Œæ‰‹å†™ã€‚å®é™…å¯ä»¥ä¸ç”¨æ‰‹å†™ã€‚

             #b1 = b1 - lr * grads[1]
        #w2 = w2 - lr * grads[2]
        #b2 = b2 - lr * grads[3]
        #w3 = w3 - lr * grads[4]
        #b3 = b3 - lr * grads[5]   #ç¬¬ä¸€forå¢åŠ ä¸€ä¸ªstep
        #w1 = w1 - lr * grads[0]            #è¿™ é‡Œä¸¤ä¸ªw1ä¸ºä¸¤ä¸ªå¯¹è±¡ã€‚åŸæ¥çš„w1å‡å»è¿™ä¸ªå€¼èµ‹å€¼ç»™ä¸€ä¸ªæ–°çš„å¯¹è±¡.w1åŸæ¥æ˜¯
                                            #tf.Variable,æ›´æ–°ä¸€æ¬¡ä¹‹åæ–°çš„w1å˜ä¸ºtf.Tensorç±»å‹äº†ã€‚æ–°ä¸€æ¬¡æ“ä½œä¹‹åå°±ä¼šé”™è¯¯ã€‚
        w1.assign_sub(lr * grads[0])       #åŸåœ°æ›´æ–°,æ•°æ®ç±»å‹ä¿æŒä¸å˜
        b1.assign_sub(lr * grads[1])
        w2.assign_sub(lr * grads[2])
        b2.assign_sub(lr * grads[3])
        w3.assign_sub(lr * grads[4])
        b3.assign_sub(lr * grads[5])

        # print(isinstance(b3,tf.Variable))
        # print(isinstance(b3,tf.Tensor))
        #æ¯100è®ºï¼Œçœ‹ä¸€ä¸‹lossä¿¡æ¯ã€‚
        if step % 100 == 0:
            print(epoch, step, 'loss:', float(loss))
```
