è¯»è€…åœ¨çœ‹åˆ°æ¯ç§å¼ é‡æ—¶ï¼Œèƒ½å¤Ÿç›´è§‚åœ°è”æƒ³åˆ°å®ƒä¸»è¦çš„ç‰©ç†æ„ä¹‰å’Œç”¨é€”ï¼Œå¯¹åç»­å¼ é‡çš„ç»´åº¦

### æ ‡é‡
åœ¨ TensorFlow ä¸­ï¼Œæ ‡é‡æœ€å®¹æ˜“ç†è§£ï¼Œå®ƒå°±æ˜¯ä¸€ä¸ªç®€å•çš„æ•°å­—ï¼Œç»´åº¦æ•°ä¸º0ï¼Œshape ä¸º
[]ã€‚æ ‡é‡çš„å…¸å‹ç”¨é€”ä¹‹ä¸€æ˜¯è¯¯å·®å€¼çš„è¡¨ç¤ºã€å„ç§æµ‹é‡æŒ‡æ ‡çš„è¡¨ç¤ºï¼Œæ¯”å¦‚å‡†ç¡®åº¦(Accuracy,
acc)ï¼Œç²¾åº¦(Precision)å’Œå¬å›ç‡(Recall)ç­‰ã€‚
è€ƒè™‘æŸä¸ªæ¨¡å‹çš„è®­ç»ƒæ›²çº¿ï¼Œå¦‚å›¾ 4.1 æ‰€ç¤ºï¼Œæ¨ªåæ ‡ä¸ºè®­ç»ƒBatch æ­¥æ•°Stepï¼Œçºµåæ ‡åˆ†
åˆ«ä¸ºè¯¯å·®å˜åŒ–è¶‹åŠ¿(å›¾ 4.1(a))å’Œå‡†ç¡®åº¦å˜åŒ–è¶‹åŠ¿æ›²çº¿(å›¾ 4.1(b))ï¼Œå…¶ä¸­æŸå¤±å€¼loss å’Œå‡†ç¡®åº¦
å‡ç”±å¼ é‡è®¡ç®—äº§ç”Ÿï¼Œç±»å‹ä¸ºæ ‡é‡ã€‚

ä»¥å‡æ–¹å·®è¯¯å·®å‡½æ•°ä¸ºä¾‹ï¼Œç»è¿‡tf.keras.losses.mse(æˆ–tf.keras.losses.MSE)è¿”å›æ¯ä¸ªæ ·æœ¬
ä¸Šçš„è¯¯å·®å€¼ï¼Œæœ€åå–è¯¯å·®çš„å‡å€¼ä½œä¸ºå½“å‰batch çš„è¯¯å·®ï¼Œå®ƒæ˜¯ä¸€ä¸ªæ ‡é‡ï¼š

```python
out = tf.random.uniform([4,10]) #éšæœºæ¨¡æ‹Ÿç½‘ç»œè¾“å‡º
y = tf.constant([2,3,2,0]) # éšæœºæ„é€ æ ·æœ¬çœŸå®æ ‡ç­¾
y = tf.one_hot(y, depth=10) # one-hot ç¼–ç 
loss = tf.keras.losses.mse(y, out) # è®¡ç®—æ¯ä¸ªæ ·æœ¬çš„MSE
loss = tf.reduce_mean(loss) # å¹³å‡MSE
print(loss)
```
